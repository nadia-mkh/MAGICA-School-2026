## Web Applications for Scientific Data

Modern scientific data workflows increasingly rely on web applications to make large, cloud-hosted datasets accessible and explorable without requiring users to download data locally. This lecture covers three complementary approaches:

### 1. Apps with Bokeh / Django Backends

These are full-stack web applications where a live Python process runs on a server, performing computation and serving interactive visualizations to the browser. The backend retains access to the full Python scientific stack (xarray, numpy, dask, etc.).

- **[Bokeh](https://bokeh.org/)** enables interactive plots that communicate with a Python server via WebSockets, allowing widgets to trigger real server-side computation.
- **[Panel](https://panel.holoviz.org/)** (built on Bokeh) makes it easy to turn notebooks and HoloViz plots into deployable dashboards.
- **[Django](https://www.djangoproject.com/)** provides a full web framework for building data portals with authentication, REST APIs, and database backends.

**Demo:** [SHYFEM model viewer](http://161.9.255.154/interactive/) — interactive ocean model output served via a Bokeh backend.

**Tradeoff:** Requires a persistent server process — more infrastructure to manage, but full Python flexibility.

---

### 2. Apps that Consume xpublish-based Web Services

[xpublish](https://xpublish.readthedocs.io/) is a Python library that serves xarray datasets over HTTP using a plugin-based architecture. It supports standardized geospatial APIs, meaning any compliant client (GIS tools, web maps, other Python code) can consume the data without knowing anything about the underlying storage format.

Key xpublish plugins and the OGC APIs they implement:

- **[OGC API – Tiles](https://ogcapi.ogc.org/tiles/)**: Serves raster or vector data as map tiles (e.g., for Leaflet/MapLibre web maps). Enables streaming of large gridded datasets to a browser at any zoom level.
- **[OGC API – Environmental Data Retrieval (EDR)](https://ogcapi.ogc.org/edr/)**: Provides queries for point, area, trajectory, and corridor extractions from gridded or in-situ datasets. Ideal for serving model output or observational data to analysis clients.

**Demo:** [Weather analysis time series app](https://opensciencecomputing.github.io/xpublish-tiles/examples/maplibre/timeseries-explorer.html) — consumes OGC API Tiles/EDR to extract and plot time series from gridded weather data.

**Tradeoff:** Lightweight and standards-based — clients don't need Python, but the server still needs xpublish running.

---

### 3. Apps that Directly Access Icechunk Repos

The most cloud-native approach: the web application reads data directly from an [Icechunk](https://icechunk.io/) repository in object storage, with no intermediary server. This is possible because Icechunk repos are Zarr stores with transactional versioning, and the Zarr metadata is structured so that clients can make efficient range requests directly to S3 (or compatible storage).

- **[zarr-layer](https://github.com/developmentseed/zarr-layer)** is a JavaScript/WebGL library that reads Zarr stores directly in the browser and renders them as map layers — no Python server required at read time. 
- This approach enables truly serverless data visualization: once data is written to an Icechunk repo, it can be visualized by any zarr-aware client.

**Demo:** [Dynamical.org Weather browser](https://opensciencecomputing.github.io/xpublish-tiles/examples/maplibre/zarr-direct.html) — accesses weather data directly from Native Icechunks in the browser using zarr-layer.

[Dynamical.org Weather browser](https://opensciencecomputing.github.io/NCICS-2026/zarr-layer-demo/) — accesses CDR-NDVI data from a Hybrid Icechunk store (virtual level 0 accessing NetCDF files, native higher levels) in the browser using zarr-layer.

**Tradeoff:** Requires data to already be in Zarr/Icechunk format, and the client must handle chunked reads efficiently.
