# Marine Heatwave Detection — Gulf of Tunis (2016–2025)

This notebook detects and analyses **marine heatwave (MHW) events** in the Gulf of Tunis over ten summer seasons (JAS: July–August–September, 2016–2025).

**Definition used (Hobday et al. 2016):** a pixel is classified as *in heatwave* on day *t* if its sea surface temperature has been ≥ the local **90th-percentile climatological threshold** for at least 5 consecutive days ending on day *t* (i.e., the 5-day trailing rolling minimum ≥ threshold).

The threshold is derived from a **fixed climatological baseline (JAS 1994–2015)** — independent of the analysis period — making the detection statistically robust and comparable across years.

| Step | Description |
|------|-------------|
| 1 | Load daily SST: climatology baseline (JAS 1994–2015) and analysis period (JAS 2016–2025) |
| 2 | Compute 90th-percentile threshold map per pixel from the 1994–2015 climatology |
| 3 | Detect heatwave pixels in 2016–2025 — rolling-minimum approach |
| 4 | Label and characterise events |
| 5 | Spatial map of MHW frequency |
| 6 | Interactive statistics plots (hvplot) |
| 7 | SST map of the largest event (peak day) |
| 8 | Chlorophyll-a map for the same date (Sentinel-3 OLCI via Planetary Computer) |

---
## Setup

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

import numpy as np
import pandas as pd
import xarray as xr
import fsspec
import copernicusmarine
import pystac_client
import planetary_computer
import hvplot.xarray
import hvplot.pandas
import panel as pn
pn.extension()

# Area of interest: Gulf of Tunis  [lon_min, lat_min, lon_max, lat_max]
bbox = [10.0, 36.6, 10.9, 37.2]

# Marine heatwave parameters (Hobday et al. 2016)
PERCENTILE = 90        # threshold percentile per pixel
MIN_DAYS   = 5         # consecutive days above threshold
MONTHS     = [7, 8, 9]              # JAS

# Single dataset covering the full period
DATASET_ID = "SST_MED_SST_L3S_NRT_OBSERVATIONS_010_012_b"
START_YEAR = 2008      # initial Copernicus period
CLIM_END   = 2015      # last year of climatology baseline  (2008–2015)
YEARS      = list(range(2016, 2026))  # analysis period

print(f"Gulf of Tunis bbox  : lon [{bbox[0]}, {bbox[2]}]  lat [{bbox[1]}, {bbox[3]}]")
print(f"MHW threshold       : {PERCENTILE}th percentile per pixel (Hobday et al. 2016)")
print(f"Min duration        : >= {MIN_DAYS} consecutive days")
print(f"Climatology baseline: JAS  {START_YEAR}–{CLIM_END}")
print(f"Analysis period     : JAS  {YEARS[0]}–{YEARS[-1]}")
```

---
## 1. Load CMEMS SST (JAS 2008–2025)

We download the full period in a single request from the **Mediterranean Sea SST L3S NRT** product (~0.01° daily), then split into baseline and analysis periods in memory.

- **Dataset ID:** `SST_MED_SST_L3S_NRT_OBSERVATIONS_010_012_b`
- **Variable:** `adjusted_sea_surface_temperature` (Kelvin)
- **Climatology baseline:** JAS 2008–2015 → used to compute the 90th-percentile threshold
- **Analysis period:** JAS 2016–2025 → heatwave detection

```python
VAR_NAME = "adjusted_sea_surface_temperature"   # Kelvin

ds_sst = copernicusmarine.open_dataset(
    DATASET_ID,
    variables=[VAR_NAME],
    minimum_longitude=bbox[0],
    maximum_longitude=bbox[2],
    minimum_latitude=bbox[1],
    maximum_latitude=bbox[3],
    start_datetime=f"{START_YEAR}-07-01T00:00:00",
    end_datetime="2025-09-30T23:59:59",
)
print(f"Dataset dims : {dict(ds_sst.sizes)}")
print(f"Time range   : {str(ds_sst.time.values[0])[:10]} -> {str(ds_sst.time.values[-1])[:10]}")
```

```python
# Keep JAS months only; convert Kelvin to Celsius
sst_raw  = ds_sst[VAR_NAME]
jas_mask = sst_raw.time.dt.month.isin(MONTHS).values
sst_jas  = (sst_raw.isel(time=jas_mask) - 273.15).rename("sst")

print(f"Loading {jas_mask.sum()} JAS days ...")
sst_jas = sst_jas.load()

# Split into climatology baseline and analysis period
sst_clim = sst_jas.sel(time=sst_jas.time.dt.year <= CLIM_END)
sst_anl  = sst_jas.sel(time=sst_jas.time.dt.year >= YEARS[0])

print(f"\nClimatology SST  : {float(sst_clim.min()):.1f} – {float(sst_clim.max()):.1f} °C"
      f"  ({str(sst_clim.time.values[0])[:10]} → {str(sst_clim.time.values[-1])[:10]})"
      f"  [{len(sst_clim.time)} days]")
print(f"Analysis SST     : {float(sst_anl.min()):.1f} – {float(sst_anl.max()):.1f} °C"
      f"  ({str(sst_anl.time.values[0])[:10]} → {str(sst_anl.time.values[-1])[:10]})"
      f"  [{len(sst_anl.time)} days]")
```

---
## 2. Marine Heatwave Detection

Following **Hobday et al. (2016)**, the threshold is the **90th percentile of SST computed per pixel** from the JAS **1994–2015 climatology** — a fixed baseline independent of the analysis period.

A pixel is *in heatwave* on analysis day *t* if its 5-day trailing rolling minimum ≥ the local threshold, meaning it has been above threshold for at least 5 consecutive days.

Rolling is computed **per year** to avoid spurious multi-year runs across the September→July season boundary.

> **Reference:** Hobday, A.J. et al. (2016). A hierarchical approach to defining marine heatwaves. *Progress in Oceanography*, 141, 227–238.

```python
# 90th-percentile threshold per pixel from the 2008–2015 climatology (Hobday et al. 2016)
threshold_map = sst_clim.quantile(PERCENTILE / 100, dim="time").drop_vars("quantile")

print(f"Threshold range : {float(threshold_map.min()):.2f} – {float(threshold_map.max()):.2f} °C")
print(f"Threshold mean  : {float(threshold_map.mean()):.2f} °C")

threshold_map.hvplot(
    x="longitude", y="latitude",
    rasterize=True, geo=True, tiles="OSM",
    cmap="RdYlBu_r",
    clabel=f"SST {PERCENTILE}th percentile (°C)",
    title=f"MHW Threshold Map ({PERCENTILE}th percentile, JAS {START_YEAR}–{CLIM_END} climatology)",
    frame_width=700, alpha=0.85,
)
```

```python
# Detect heatwave days in 2016–2025 against the 2008–2015 climatological threshold
hw_list = []
for year in YEARS:
    yr_mask = sst_anl.time.dt.year.values == year
    sst_y   = sst_anl.isel(time=yr_mask)
    if len(sst_y.time) == 0:
        continue
    # 5-day trailing rolling minimum (NaN where fewer than 5 valid days)
    roll_min = sst_y.rolling(time=MIN_DAYS, min_periods=MIN_DAYS).min()
    # Compare against spatially varying 2008–2015 climatological threshold
    hw       = (roll_min >= threshold_map).fillna(False)
    hw_list.append(hw)

heatwave = xr.concat(hw_list, dim="time")   # (n_days, lat, lon) bool
print(f"Heatwave mask shape    : {heatwave.shape}")
print(f"Pixel-days in heatwave : {float(heatwave.mean()):.3%}")
```

### 2b. Event labelling

A **basin-scale event** is a contiguous sequence of days on which at least one pixel is in heatwave.  Events are numbered 1, 2, 3, … in chronological order; year boundaries are treated as hard breaks.

```python
# Domain-wide pixel count per day
n_hw   = heatwave.sum(dim=["latitude", "longitude"]).values.astype(float)
times  = pd.DatetimeIndex(heatwave.time.values)
active = n_hw > 0

event_id      = np.zeros(len(times), dtype=int)
current_event = 0
in_evt        = False

for i in range(len(times)):
    if i > 0 and times[i].year != times[i - 1].year:   # hard year break
        in_evt = False
    if active[i]:
        if not in_evt:
            current_event += 1
            in_evt = True
        event_id[i] = current_event
    else:
        in_evt = False

n_events = current_event
print(f"Total heatwave events detected : {n_events}")
```

```python
rows = []
for evt in range(1, n_events + 1):
    mask     = event_id == evt
    peak_idx = int(np.argmax(n_hw * mask))   # day with most heatwave pixels
    rows.append({
        "event_id"   : evt,
        "year"       : int(times[mask][0].year),
        "start"      : times[mask][0],
        "end"        : times[mask][-1],
        "duration"   : int(mask.sum())+4,
        "peak_date"  : times[peak_idx],
        "max_pixels" : int(n_hw[mask].max()),
        "mean_pixels": float(n_hw[mask].mean()),
    })

events_df = pd.DataFrame(rows)
print(events_df.to_string(index=False))
```

### 2.c SST Map — Heaviest Heatwave Event (Peak Day)

We select the peak day of the largest event (by max heatwave pixels) and show:
- the full SST field everywhere (background, `RdYlBu_r`)
- the heatwave boundary as a dark red contour line

```python
event_max = events_df[events_df["event_id"] == 9].iloc[0]
peak_date = event_max["peak_date"]

print(f"Largest event : #{int(event_max['event_id'])}  {event_max['start'].date()} → {event_max['end'].date()}")
print(f"Duration      : {event_max['duration']} days")
print(f"Peak day      : {peak_date.date()}  ({int(event_max['max_pixels'])} heatwave pixels)")
```

```python
sst_peak = sst_anl.sel(time=peak_date, method="nearest")
hw_peak  = heatwave.sel(time=peak_date, method="nearest")

# Full SST field as background
sst_map = sst_peak.hvplot(
    x="longitude", y="latitude",
    rasterize=True, geo=True, tiles="OSM",
    cmap="RdYlBu_r",
    clabel="SST (°C)",
    title=f"SST — Largest MHW Event Peak Day ({peak_date.date()})",
    frame_width=700,
)

# Heatwave boundary as a single dark red contour
hw_boundary = hw_peak.astype(float).hvplot.contour(
    x="longitude", y="latitude",
    levels=[0.5],
    line_color="#8B0000",
    line_width=2,
    geo=True,
)

sst_map * hw_boundary
```

### 2d. Spatial Map of MHW Frequency

For each pixel, we compute the fraction of JAS days **2016–2025** it was classified as *in heatwave* (threshold from the 1994–2015 climatology).  Warmer colours indicate pixels that are persistently affected by marine heatwaves.

```python
import matplotlib.cm as cm

# Sea mask: pixels that have at least one valid SST value (excludes land)
sea_mask = sst_jas.notnull().any(dim="time")

# Fraction of JAS days each ocean pixel was in heatwave (2016–2025), in percent
mhw_freq = heatwave.mean(dim="time") * 100
mhw_freq = mhw_freq.where(sea_mask)

mhw_freq.hvplot(
    x="longitude", y="latitude",
    rasterize=True, geo=True, tiles="OSM",
    cmap = cm.get_cmap("RdYlBu").reversed(),
    clim=(0, float(mhw_freq.quantile(0.99))),
    clabel="MHW frequency (%)",
    title="Marine Heatwave Frequency — Gulf of Tunis (JAS 2016–2025)",
    frame_width=700,
)
```

## 3. Statistics and Visualisations

### 3a. Number of events per year

```python
events_per_year = (
    events_df.groupby("year")
    .size()
    .reindex(YEARS, fill_value=0)
    .reset_index(name="n_events")
)

events_per_year.hvplot.bar(
    x="year", y="n_events",
    title="Marine Heatwave Events per Year — Gulf of Tunis (JAS)",
    xlabel="Year", ylabel="Number of Events",
    color="coral",
    frame_width=700, 
)
```

### 3b. Max heatwave pixels per event (coloured by year)

```python
import matplotlib.pyplot as plt

total_sea_pixels = int(sea_mask.sum())
events_df["max_pixels_pct"] = (events_df["max_pixels"] / total_sea_pixels * 100).round(2)

years  = sorted(events_df["year"].unique())
colors = {yr: c for yr, c in zip(years, plt.cm.tab10.colors)}

fig, ax = plt.subplots(figsize=(12, 5))

bar_w   = 0.7
year_gap = 1.2
current_x = 0
year_tick_positions = []

for year in years:
    evts = events_df[events_df["year"] == year].reset_index(drop=True)
    positions = [current_x + i * bar_w for i in range(len(evts))]
    ax.bar(positions, evts["max_pixels_pct"], width=bar_w * 0.9,
           color=colors[year], label=str(year))
    year_tick_positions.append((current_x + (len(evts) - 1) * bar_w / 2, str(year)))
    current_x += len(evts) * bar_w + year_gap

ax.set_xticks([x for x, _ in year_tick_positions])
ax.set_xticklabels([l for _, l in year_tick_positions])
ax.set_title("Area% per Event — Gulf of Tunis (JAS 2016–2025)")
ax.set_xlabel("Year")
ax.set_ylabel("Percentage of area with heatwave")
ax.legend(title="Year")
plt.tight_layout()
plt.show()
```

## 4. Chlorophyll-a Map — Same Date (Sentinel-3A OLCI)

We search the **Planetary Computer STAC API** for a Sentinel-3 OLCI WFR L2 pass closest to the event-3 peak day, then apply the same spatial masking used in the Gulf of Tunis tutorial.

- **Collection:** `sentinel-3-olci-wfr-l2-netcdf`
- **Variable:** `CHL_NN` (Neural Net chlorophyll, mg m⁻³) — best for coastal/turbid waters
- **Resolution:** ~300 m

Lat/lon coordinates live in a separate `geo-coordinates` asset and are merged before masking.

```python
catalog = pystac_client.Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

# Search +/-7 days around peak to maximise chance of a cloud-free pass
t0 = (peak_date - pd.Timedelta(days=7)).date()
t1 = (peak_date + pd.Timedelta(days=7)).date()

search_chl = catalog.search(
    collections=["sentinel-3-olci-wfr-l2-netcdf"],
    bbox=bbox,
    datetime=f"{t0}/{t1}",
)
items_chl = search_chl.item_collection()
print(f"Found {len(items_chl)} OLCI passes in [{t0}, {t1}]")
for it in items_chl:
    print(f"  {it.id}  ({it.datetime.date()})")
```

```python
from datetime import timezone

# Pick the pass closest to the peak day
peak_utc = peak_date.to_pydatetime().replace(tzinfo=timezone.utc)
item_chl  = min(items_chl, key=lambda x: abs((x.datetime - peak_utc).total_seconds()))
print(f"Selected pass : {item_chl.id}")
print(f"Pass datetime : {item_chl.datetime}")
```

```python
def open_asset(item, key):
    """Sign and open a Planetary Computer asset as an xarray Dataset."""
    asset = planetary_computer.sign(item.assets[key])
    return xr.open_dataset(fsspec.open(asset.href).open())

ds_chl = open_asset(item_chl, "chl-nn")
geo    = open_asset(item_chl, "geo-coordinates")

# Merge lat/lon (stored in a separate asset) as 2-D coordinates
ds_chl = ds_chl.assign_coords(
    latitude= (["rows", "columns"], geo["latitude"].values),
    longitude=(["rows", "columns"], geo["longitude"].values),
)
print(ds_chl)
```

```python
lat_chl = ds_chl["latitude"].values
lon_chl = ds_chl["longitude"].values

mask_chl = (
    (lat_chl >= bbox[1]) & (lat_chl <= bbox[3]) &
    (lon_chl >= bbox[0]) & (lon_chl <= bbox[2])
)

rows_idx, cols_idx = np.where(mask_chl)
print(f"{len(rows_idx)} OLCI pixels inside Gulf of Tunis bbox")

row_min, row_max = rows_idx.min(), rows_idx.max() + 1
col_min, col_max = cols_idx.min(), cols_idx.max() + 1

chl_sub = ds_chl["CHL_NN"].isel(
    rows=slice(row_min, row_max),
    columns=slice(col_min, col_max),
).load()

lat_sub  = lat_chl[row_min:row_max, col_min:col_max]
lon_sub  = lon_chl[row_min:row_max, col_min:col_max]
mask_sub = mask_chl[row_min:row_max, col_min:col_max]

chl_vals = chl_sub.values.copy().astype(float)
chl_vals[~mask_sub] = np.nan

valid_chl = chl_vals[np.isfinite(chl_vals) & (chl_vals > 0)]
print(f"CHL_NN range : {valid_chl.min():.3f} - {valid_chl.max():.3f} mg m-3  |  mean: {valid_chl.mean():.3f}")
```

```python
df_chl = pd.DataFrame({
    "lat": lat_sub.flatten(),
    "lon": lon_sub.flatten(),
    "chl": chl_vals.flatten(),
})
df_chl = df_chl[df_chl["chl"].notna() & (df_chl["chl"] > 0) & (df_chl["chl"] < 100)]
df_chl["log_chl"] = np.log10(df_chl["chl"])

df_chl.hvplot.points(
    x="lon", y="lat",
    c="log_chl",
    colormap="YlGn",
    clabel="log10 CHL (mg m-3)",
    title=f"Sentinel-3 OLCI Chlorophyll-a — Gulf of Tunis ({item_chl.datetime.date()})",
    xlabel="Longitude", ylabel="Latitude",
    size=6,
    geo=True,
    tiles="OSM",
    frame_width=700,
)
```

## 5. Summary

| Step | Dataset | Key variable | Resolution |
|------|---------|-------------|------------|
| Climatology threshold | CMEMS MED SST L3S MY (reprocessed) | `adjusted_sea_surface_temperature` | ~0.01° daily |
| MHW detection | CMEMS MED SST L3S NRT | `adjusted_sea_surface_temperature` | ~0.01° daily |
| Chlorophyll-a | Sentinel-3 OLCI WFR L2 (Planetary Computer) | `CHL_NN` | ~300 m |

**Key points:**
- The threshold follows **Hobday et al. (2016)**: the 90th percentile per pixel computed from a fixed **1994–2015 baseline**, making it independent of the warming trend observed during 2016–2025.  Using the analysis period itself as the baseline would systematically undercount recent events.
- The **rolling-minimum** approach ensures only *sustained* exceedances (≥ 5 consecutive days) are flagged, filtering out brief warm spikes.
- Chlorophyll-a is plotted on a **log₁₀ scale** because coastal values span several orders of magnitude.
- Elevated chlorophyll near the Medjerda River delta in summer often co-occurs with SST anomalies — the spatial patterns can hint at upwelling or nutrient loading.
